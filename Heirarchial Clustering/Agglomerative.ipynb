```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
iris=load_iris()
iris_data=pd.DataFrame(iris.data)
iris_data.columns=iris.feature_names
iris_data
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X=scaler.fit_transform(iris_data)
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
X_scaled=pca.fit_transform(X)
```

```python
plt.scatter(X_scaled[:,0],X_scaled[:,1],c=iris.target)
```

```python
import scipy.cluster.hierarchy as sc
plt.figure(figsize=(20,7))
plt.title("Dendrogram")
sc.dendrogram(sc.linkage(X_scaled,method='ward'))
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster=AgglomerativeClustering(n_clusters=2,linkage='ward')
cluster.fit(X_scaled)
```

```python
cluster.labels_
```

```python
plt.scatter(X_scaled[:,0],X_scaled[:,1],c=cluster.labels_)
```
